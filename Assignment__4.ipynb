```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/NCI60_X.csv",index_col = 0)
```

```python
import seaborn as sns
```

```python
from sklearn.preprocessing import scale
X = pd.DataFrame(scale(data), index = data.index, columns = data.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components= 2)
```

```python
pc_array = pc.fit_transform(X)
```

```python
pc_array.shape
```

```python
pc_data = pd.DataFrame(pc_array,columns = ["pc1","pc2"])
```

```python
pc_data.shape
```

```python
pc_data
```

```python
pc_data.corr().round(1)
```

```python
pc.explained_variance_
```

```python
pc.explained_variance_ratio_ * 100
```

```python
plt.plot([1,2],pc.explained_variance_ratio_ * 100,"-o")
plt.show()
```

```python
pc = PCA(n_components = 20)
```

```python
pc_array = pc.fit_transform(X)
```
